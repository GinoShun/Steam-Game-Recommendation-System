# Model and Performance visualization

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
```

```python
# Function to add labeled rectangles to represent the layers, with vertical text
def add_layer_vertical(ax, text, xy, width=0.8, height=3, color="skyblue"):
    box = FancyBboxPatch(xy, width, height, boxstyle="round,pad=0.1", edgecolor="black", facecolor=color)
    ax.add_patch(box)
    ax.text(xy[0] + width / 2, xy[1] + height / 2, text, ha="center", va="center", fontsize=12, rotation=90, wrap=True)
```

#### LSTM

```python

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Hide axes
ax.axis('off')

# Define positions, sizes, and colors for each layer in a sequential layout with shades of blue, all centered horizontally
layer_details_final = [
    ("Input\n(Sequence Length: 3)", (-6, -2.5), 0.8, 5, "#6baed6"),          # Medium blue
    ("Masking Layer\n(Mask Value: -1)", (-4, -2.5), 0.8, 5, "#b2d7e9"),       # Slightly darker blue
    ("LSTM Layer 1\n(Units: 256)", (-2, -4), 0.8, 8, "#87bdd8"),              # Darker blue
    ("Dropout Layer\n(Rate: 0.5)", (0, -2.5), 0.8, 5, "#aec6cf"),             # Light blue
    ("LSTM Layer 2\n(Units: 128)", (2, -3.5), 0.8, 7, "#87bdd8"),             # Slightly darker blue
    ("Dropout Layer\n(Rate: 0.5)", (4, -2.5), 0.8, 5, "#aec6cf"),             # Light blue
    ("Dense Layer 1\n(Units: 64)", (6, -3), 0.8, 6, "#9ecae1"),               # Darker blue
    ("Output Layer\n(Units: 1)", (8, -1.5), 0.8, 3, "#6baed6")                # Medium blue
]

# Add all the layers to the diagram with all blocks centered horizontally
for text, position, width, height, color in layer_details_final:
    add_layer_vertical(ax, text, position, width, height, color)

# Add straight arrows between layers, horizontally centered
for i in range(len(layer_details_final) - 1):
    start_x = layer_details_final[i][1][0] + 0.8
    end_x = layer_details_final[i + 1][1][0]
    ax.annotate('', xy=(end_x, 0), xytext=(start_x, 0),
                arrowprops=dict(arrowstyle='->', lw=1.5))

# Set plot limits
ax.set_xlim(-7, 10)
ax.set_ylim(-5, 5)

# Show the diagram
plt.tight_layout()
plt.savefig('../data/plots/lstm_model.png')
plt.show()

```

#### NCF

```python
# Create figure and axis for the NCF model diagram
fig, ax = plt.subplots(figsize=(14, 10))

# Hide axes
ax.axis('off')

# Define positions, sizes, and colors for each layer in the NCF model, all centered horizontally, with proportional heights
ncf_layer_details = [
    ("User Input\n(Input Shape: 1)", (-8, -1), 0.8, 2, "#aec6cf"),                # Light blue (Input Layer)
    ("Item Input\n(Input Shape: 1)", (-6, -1), 0.8, 2, "#aec6cf"),                # Light blue (Input Layer)
    ("User Embedding\n(Output Shape: (1, 8))", (-4, -3), 0.8, 6, "#87bdd8"),       # Darker blue (Embedding Layer)
    ("Item Embedding\n(Output Shape: (1, 8))", (-2, -3), 0.8, 6, "#87bdd8"),       # Darker blue (Embedding Layer)
    ("Flatten User\n(Output Shape: 8)", (0, -1.5), 0.8, 3, "#084594"),             # Steel blue (Flatten Layer)
    ("Flatten Item\n(Output Shape: 8)", (2, -1.5), 0.8, 3, "#084594"),             # Steel blue (Flatten Layer)
    ("Features Input\n(Input Shape: 20)", (4, -2), 0.8, 4, "#c6dbef"),             # Pale blue (Input Feature Layer)
    ("Concatenate\n(Output Shape: 36)", (6, -3), 0.8, 6, "#9ecae1"),               # Light sky blue (Concatenate Layer)
    ("Dense 1\n(Output Shape: 256)", (8, -8), 0.8, 16, "#6baed6"),                 # Medium blue (Dense Layer)
    ("Dropout 1\n(Output Shape: 256)", (10, -8), 0.8, 16, "#aec6cf"),              # Light blue (Dropout Layer)
    ("Dense 2\n(Output Shape: 128)", (12, -6), 0.8, 12, "#6baed6"),                # Medium blue (Dense Layer)
    ("Dropout 2\n(Output Shape: 128)", (14, -6), 0.8, 12, "#aec6cf"),              # Light blue (Dropout Layer)
    ("Dense 3\n(Output Shape: 64)", (16, -4), 0.8, 8, "#6baed6"),                  # Medium blue (Dense Layer)
    ("Dense 4\n(Output Shape: 32)", (18, -3), 0.8, 6, "#6baed6"),                  # Medium blue (Dense Layer)
    ("Output\n(Output Shape: 1)", (20, -1), 0.8, 2, "#aec6cf")                     # Dark blue (Output Layer)
]

# Add all the layers to the diagram with all blocks centered horizontally
for text, position, width, height, color in ncf_layer_details:
    add_layer_vertical(ax, text, position, width, height, color)

# Add straight arrows between layers, horizontally centered
for i in range(len(ncf_layer_details) - 1):
    start_x = ncf_layer_details[i][1][0] + 0.8
    end_x = ncf_layer_details[i + 1][1][0]
    ax.annotate('', xy=(end_x, 0), xytext=(start_x, 0),
                arrowprops=dict(arrowstyle='->', lw=1.5))

# Set plot limits
ax.set_xlim(-10, 22)
ax.set_ylim(-9, 7)

# Show the diagram
plt.tight_layout()
plt.show()
```
